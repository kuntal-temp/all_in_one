```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

# Dataset

```python
penguins = sb.load_dataset("penguins")
penguins.head()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
flights = sb.load_dataset("flights")
flights_may = flights.query("month == 'May'")
flights.head()
```

```python
tips = sb.load_dataset("tips")
tips.head()
```

```python
tips['time'].value_counts()
```

```python
tips['day'].value_counts()
```

```python
iris = sb.load_dataset("iris")
iris.head()
```

```python
exercise = sb.load_dataset("exercise")
exercise.head()
```

```python
df = pd.read_csv("../test_data/wine.csv")
df.head()
```

# lineplot()

Relational plots

```python
sb.lineplot(x="month", y="passengers", data=flights)
```

```python
sb.lineplot(x="month", y="passengers", hue='year', data=flights)
```

```python
sb.lineplot(x="year", y="passengers", data=flights_may)
```

# relplot()

Relational plots

```python
sb.relplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sb.relplot(data=tips, x="total_bill", y="tip", hue="sex", kind="scatter")
```

```python
sb.relplot(data=tips, x="total_bill", y="tip", hue="sex", kind="line")
```

```python
sb.relplot(data=tips, x="total_bill", y="tip", hue="day", col="time")
```

```python
sb.relplot(data=tips, x="total_bill", y="tip", hue="day", row="sex")
```

```python
sb.relplot(data=tips, x="total_bill", y="tip", hue="day", col="time", row="sex")
```
