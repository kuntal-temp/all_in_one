```python
x = np.arange(0,10,1)
y = x**2
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, label='Label_1')
axes.plot(x+5, y*2, label='Label_2')
axes.plot(x+10, y*4, label='Label_3')
axes.plot(x+15, y*6, label='Label_4')
axes.plot(x+20, y*8, label='Label_5')
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title("Simple Plot")
axes.legend()
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(5,2)) # figsize=(width, hight)

axes[0].plot(x,y, 'y', label="Label 1")
axes[1].plot(x,y, 'g', label="Label 2")

axes[0].grid(True)

axes[0].legend()
axes[1].legend()

plt.show()
```

# fill_between(x, y1, y2)

```python
x = np.linspace(0, 8, 16)
y1 = 3 + 4*x/8 + np.random.uniform(0.0, 0.5, len(x))
y2 = 1 + 2*x/8 + np.random.uniform(0.0, 0.5, len(x))
```

```python
fig, ax = plt.subplots()

ax.fill_between(x, y1, y2, alpha=.4, linewidth=0)
ax.plot(x, (y1 + y2)/2, linewidth=2)

plt.show()
```

# Statistics plots

```python
# hist(x)
x = 4 + np.random.normal(0, 1.5, 200)

plt.hist(x, bins=8)

plt.show()
```

```python
# hist2d(x, y)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3
fig, ax = plt.subplots()

ax.hist2d(x, y)

plt.show()
```

```python
# boxplot(X)
D = np.random.normal((3, 5, 4), (1.25, 1.00, 1.25), (100, 3))
fig, ax = plt.subplots()

VP = ax.boxplot(D, positions=[2, 4, 6], widths=1.5)

plt.show()
```

```python
fig, ax = plt.subplots()

VP = ax.boxplot(D, 
                positions=[2, 4, 6], 
                widths=1.5, 
                patch_artist=True,
                showmeans=True,
                medianprops={"color": "white", "linewidth": 0.5},
                boxprops={"facecolor": "C0", "edgecolor": "white", "linewidth": 0.5},
                whiskerprops={"color": "C0", "linewidth": 1.5},
                capprops={"color": "C0", "linewidth": 1.5}
)

plt.show()
```

```python
# errorbar(x, y, yerr, xerr)
x = [2, 4, 6]
y = [3.6, 5, 4.2]
yerr = [0.9, 1.2, 0.5]
xerr = yerr = [0.9, 1.2, 0.5]
fig, ax = plt.subplots()

ax.errorbar(x, y, yerr, xerr, fmt='o', linewidth=2, capsize=6)

ax.grid(True)
plt.show()
```

```python
# violinplot(D)
D = np.random.normal((3, 5, 4), (0.75, 1.00, 0.75), (200, 3))
fig, ax = plt.subplots()

vp = ax.violinplot(D, 
                   [2, 4, 6], 
                   widths=2,
                   showmeans=True, 
                   showmedians=True, 
                   showextrema=True
)

plt.show()
```

```python
# hexbin(x, y, C)
x = np.random.randn(5000)
y = 1.2 * x + np.random.randn(5000) / 3
fig, ax = plt.subplots()

ax.hexbin(x, y, gridsize=20)

plt.show()
```

```python
# pie(x)
x = [1, 2, 3, 4]
fig, ax = plt.subplots()

ax.pie(x, 
       radius=50, 
       center=(4, 4),
       wedgeprops={"linewidth": 1, "edgecolor": "white"}, 
       frame=True)

plt.show()
```
