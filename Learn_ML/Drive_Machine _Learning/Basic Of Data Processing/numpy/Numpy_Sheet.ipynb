```python
import numpy as np
from numpy import pi
from numpy import poly1d
print(np.__version__)
```

```python
# arange
array = np.arange(16)
print(array)
array = np.arange(5, dtype='complex')
print(array)
array = np.arange(5, 16)
print(array)
array = np.arange(10, 0, -4)
print(array)
array = np.arange(5, 16, 3)
print(array)
```

```python
# linspace
array = np.linspace(5, 7, 10)  # here 10 to 20 data has broken in 50 points
print("Array : ", array)
```

```python
# zeros
array = np.zeros(10)
print(array)
array = np.zeros((2, 4), dtype='complex')
print(array)
```

```python
# ones
array = np.ones(10)
print(array)
array = np.ones((2, 4), dtype='complex')
print(array)
```

```python
# empty
array = np.empty(10)
print(array)
array = np.empty((3,3), dtype='int')
print(array)
```

```python
# eye
array = np.eye(4, dtype='float')    # default N=M=4
print(array)
array = np.eye(N=3, M=3, k=0, dtype='float')  # k=0 is default
print(array)
array = np.eye(N=3, M=3, k=2, dtype='float')  # skip 2 col
print(array)
array = np.eye(N=3, M=3, k=-1, dtype='float') # except 1st row
print(array)
```

```python
# diag
array = np.diag((-1, -2, -3))
print(array)
array = np.diag((-1, -2, -3), k=2)  # skip 2 col
print(array)
array = np.diag((-1, -2, -3), k=-2) # skip 2 row
print(array)
```
