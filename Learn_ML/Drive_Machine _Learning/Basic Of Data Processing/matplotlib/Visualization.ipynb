```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
x = np.arange(0,100, 10)
y = x**5
```

# Matplotlib

```python
plt.plot(x, y)
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
# plt.xlim(0,1)
# plt.ylim(0,1)
plt.show()
```

```python
plt.scatter(x, y)
```

```python
fig = plt.figure(figsize=(5,4))
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(x, y)
axes.set_xlabel('X Axis Title Here')
axes.set_ylabel('Y Axis Title Here')
axes.set_title('String Title Here')
plt.show()
```

```python
# row 1, col 2
# fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10,4))
# axes[0].plot(x,y, 'y', label="Label 1")
# axes[1].plot(x,y, 'g', label="Label 2")
# axes[0].legend()

# row 2, col 2
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(15,8))
axes[0][0].plot(x,y, 'r')
axes[0][1].scatter(x, y)
axes[0][2].bar(x,y)
axes[0][3].stem(x, y)
axes[1][0].step(x, y)
axes[1][1].hist(x, bins=20)
axes[1][2].triplot(x, y)
```
