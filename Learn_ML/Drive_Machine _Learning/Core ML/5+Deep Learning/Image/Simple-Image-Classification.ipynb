## Import Dataset

```python
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
train_labels.shape
```

### Image Shape

```python
batch_size = 0
img_height, img_width = 32, 32
input_shape=(img_height, img_width, 3)
```

## Creating CNN Network

```python
model = models.Sequential()

# Normalize pixel values to be between 0 and 1
model.add(layers.Rescaling(1./255, input_shape=input_shape))

model.add(layers.Conv2D(128, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))  # AveragePooling2D

model.add(layers.Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2))

model.add(layers.Conv2D(32, kernel_size=(3, 3), activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2))

model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(32, activation='relu'))
model.add(layers.Dense(10, activation='softmax'))   # 10 cat to identify

model.summary()
```

```python
model.compile(
    optimizer='adam', 
    loss=tf.keras.losses.SparseCategoricalCrossentropy(), 
    metrics=['accuracy']
)

early_stop = EarlyStopping(monitor='val_loss', patience=10)

model.fit(
    train_images,
    train_labels,
    validation_data=(test_images, test_labels),
    epochs=1,
    callbacks=[early_stop]
)
```

```python
pd.DataFrame(model.history.history).plot()
```

```python
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print("test_loss = ", test_loss)
print("test_acc = ", test_acc)
```
